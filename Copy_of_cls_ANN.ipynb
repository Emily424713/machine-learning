```python
# from tensorflow import keras
# from tensorflow.keras import layers

# # set ANN model
# model = keras.Sequential([
#     layers.Dense(64, activation='relu', input_shape=(X_train.shape[1] -1,)),
#     layers.Dense(32, activation='relu'),
#     layers.Dense(1)  # Output layer for regression, no activation function
# ])

# model.compile(optimizer='adam', loss='mean_absolute_error')
# model.fit(X_train.loc[:, X_train.columns != 'ID'], y_train, epochs=50, batch_size=32, validation_split=0.3)


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam


# Normalizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Preprocess the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
def build_ann_model(input_dim):
    model = Sequential()
    model.add(Dense(128, input_dim=input_dim, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification

    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Initialize the ANN model
ann_model = build_ann_model(X_train.shape[1])

# Training the model
history = ann_model.fit(X_train, y_train, epochs=100, batch_size=10, validation_split=0.2)
```

```python
from sklearn.metrics import mean_absolute_error

# Making predictions
y_pred = ann_model.predict(X_test)
y_pred
```

```python


# Since we used a sigmoid activation function in the output layer, the predictions will be in probabilities.
# If you need binary predictions, you can convert these probabilities to 0 or 1 based on a threshold, typically 0.5.
y_pred_binary = (y_pred > 0.5).astype(int)
y_pred_binary
```

```python

# Calculating Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred_binary)

print("Mean Absolute Error:", mae)
```

```python
from keras.metrics import MeanAbsoluteError

# evaluated the model on the test set
mae = ann_model.evaluate(X_test.loc[:, X_test.columns != 'ID'], y_test)
print(f'Mean Absolute Error (MAE): {mae}')

# # predictions
# y_pred = ann_model.predict(X_test.loc[:, X_test.columns != 'ID'])
```
